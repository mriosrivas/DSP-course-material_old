# Complex IDFT

The Complex Inverse Discrete Fourier Transform is defined as:

$$x[n] = \sum\limits^{N-1}_{k=0}{X[k]e^{j}\frac{2\pi k n}{N}} $$

Where $x[n]$ has $N-1$ points.

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plt.rcParams["figure.figsize"] = (25,10)
X = np.loadtxt(fname = "dft.dat", dtype=complex).flatten()
plt.plot(np.absolute(X));
```

```python
N = len(X)
x = np.zeros(N, dtype=complex)
for n in range(N):
    for k in range(N):
        x[n] = x[n] + X[k]*np.exp(1j*2*np.pi*k*n/N)
```

```python
plt.rcParams["figure.figsize"] = (15,5)

samples = np.arange(len(x))
normalized_frequency = samples/len(x)

plt.subplot(1, 1, 1)
plt.plot(normalized_frequency, x, '.-')
plt.title('Synthesis', size = 25)
plt.xlabel('sample')


plt.show()
```
