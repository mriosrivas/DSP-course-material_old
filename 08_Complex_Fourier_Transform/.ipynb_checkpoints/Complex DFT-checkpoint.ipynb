# Complex DFT

The Complex Discrete Fourier Transform is defined as:

$$X[k] = \frac{1}{N}\sum\limits^{N-1}_{n=0}{x[n]e^{-j}\frac{2\pi k n}{N}} $$

Where $X[k]$ has $N-1$ points.

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plt.rcParams["figure.figsize"] = (25,10)
x = np.loadtxt(fname = "signal.dat").flatten()
plt.plot(x);
```

```python
N = len(x)
X = np.zeros(N, dtype=complex)
for k in range(N):
    for n in range(N):
        X[k] = X[k] + x[n]*np.exp(-1j*2*np.pi*k*n/N)
X = X/N
```

```python
plt.rcParams["figure.figsize"] = (15,5)

samples = np.arange(len(x))
normalized_frequency = samples/len(x)

plt.subplot(1, 1, 1)
plt.plot(normalized_frequency, np.absolute(X), '.-')
plt.title('Complex DFT', size = 25)
plt.xlabel('sample')


plt.show()
```

```python
np.savetxt('dft.dat', X)
```

```python
plt.rcParams["figure.figsize"] = (15,5)

samples = np.arange(len(x))
normalized_frequency = samples/len(x)
frequency = 32000

plt.subplot(2, 1, 1)
plt.plot(normalized_frequency*2*frequency, np.absolute(X), '.-')
plt.title('Amplitude')
plt.xlabel('frequency')


plt.subplot(2, 1, 2)
plt.plot(normalized_frequency*2*frequency, np.angle(X), '.-')
plt.title('Angle')
plt.xlabel('frequency');


#plt.plot(np.arctan2(np.imag(X), np.real(X))*180/np.pi)
```
