# Band-Pass Filters
A Band-Pass filter is just a combination of different filters. To generate a band pass filter we can do as follows:
1. Design two low-pass filters with frequecies $f_a$ and $f_b$. Where $f_a < f_b$.
2. Spectrally invert the low-pass filter with frequency $f_b$.
3. Add both fileter kernels.
4. Spectrally invert the filter generated in the previous step.

```python
import numpy as np
import matplotlib.pyplot as plt

def shifted_sinc_function(i,fc, M):
    """ 
    Function that calculates a sinc shifted time response.
  
    Parameters: 
    i (numpy array): Array of numbers representing the samples being used to construct the sinc response.
    fc (float): Cut-off frequency for the low-pass filter. Between 0 and 0.5.
    M (int): Length of the filter kernel. Usually M = 4/BW, where BW is the filter bandwidth of the transition band.
  
    Returns: 
    numpy array: Returns sinc shifted time domain response.
    """
    limit = np.where(i == M/2)[0][0]
    h = np.zeros(len(i))
    h[:limit] = (np.sin(2*np.pi*(i[:limit]-M/2)*fc))/((i[:limit]-M/2)*np.pi)
    h[limit+1:] = (np.sin(2*np.pi*(i[limit+1:]-M/2)*fc))/((i[limit+1:]-M/2)*np.pi)
    h[limit] = 2*fc
    return h


def blackman_window(i, M):
    """ 
    Function that calculates a Blackman window of a given M-kernel.
  
    Parameters: 
    i (numpy array): Array of numbers representing the samples being used to construct the Blackman window.
    M (int): Length of the filter kernel. Usually M = 4/BW, where BW is the filter bandwidth of the transition band.
  
    Returns: 
    numpy array: Returns Blackman window of a given M-kernel.
    """
    
    return 0.42 - 0.5*np.cos(2*np.pi*i/M) + 0.08*np.cos(4*np.pi*i/M)


def spectral_inversion(x):
    """ 
    Function that calculates the Spectral Inversion of a filter x.
  
    Parameters: 
    x (numpy array): Array of numbers representing the coeficients of a FIR filter.
  
    Returns: 
    numpy array: Returns the filter coeficients of a Spectral Inverted filter.
    """
    x_invert = -x
    center = int(len(x)/2)
    x_invert[center] = x_invert[center]+1
    return x_invert

def normalize_time_function(x):
    """ 
    Function that normalize a discrete time domain function.
  
    Parameters: 
    x (numpy array): Array of points representing a discrete time domain function.
  
    Returns: 
    numpy array: Returns a normalized discrete time domain function.
    """
    return x/np.sum(x)

def normalize_frequency_function(x):
    """ 
    Function that normalize a discrete frequency domain function.
  
    Parameters: 
    x (numpy array): Array of points representing a discrete frequency domain function.
  
    Returns: 
    numpy array: Returns a normalized discrete frequency domain function.
    """
    return np.absolute(x)/np.sum(np.absolute(x))


def plot(i, time_fun_1, time_fun_2, freq_fun_1, freq_fun_2, titles):
    """ 
    Function to make a 2-by-2 plot given 4 different inputs.
  
    Parameters: 
    i (numpy array): Array of points representing a discrete time domain interval.
    time_fun_1 (numpy array): Array of points representing a discrete time domain signal.
    time_fun_2 (numpy array): Array of points representing a discrete time domain signal.
    freq_fun_1 (numpy array): Array of points representing a discrete frequency domain signal.
    freq_fun_2 (numpy array): Array of points representing a discrete frequency domain signal.
    
    """
    plt.rcParams["figure.figsize"] = (15,10)

    plt.subplot(2,2,1)
    plt.stem(i, time_fun_1, markerfmt='.', use_line_collection=True)
    plt.title(titles['title_1'])

    plt.subplot(2,2,2)
    plt.stem(i, time_fun_2, linefmt='orange', markerfmt='r.', use_line_collection=True)
    plt.title(titles['title_2'])


    plt.subplot(2,2,3)
    plt.stem(i/np.max(i), freq_fun_1, markerfmt='.', use_line_collection=True)
    plt.title(titles['title_3'])


    plt.subplot(2,2,4)
    plt.stem(i/np.max(i), freq_fun_2, linefmt='orange', markerfmt='r.', use_line_collection=True)
    plt.title(titles['title_4']);
```

### Step 1
We set the parameters for both low pass filters.

```python
fc_1 = 0.14
fc_2 = 0.25
BW = 0.04
M = int(4/BW)
i = np.arange(0,M,1)

print("Filter lenght is {}".format(M))
```

```python
normalized_shifted_sinc_1 = normalize_time_function(shifted_sinc_function(i,fc_1, M))
normalized_shifted_sinc_2 = normalize_time_function(shifted_sinc_function(i,fc_2, M))


normalized_fft_shifted_sinc_1 = normalize_frequency_function(np.fft.fft(normalized_shifted_sinc_1))
normalized_fft_shifted_sinc_2 = normalize_frequency_function(np.fft.fft(normalized_shifted_sinc_2))

titles = {'title_1':'Sinc - Blackman Window with fc = {}'.format(fc_1),
          'title_2': 'Sinc - Blackman Window with fc = {}'.format(fc_2),
          'title_3': 'Frequency response of Low-Pass Filter with fc = {}'.format(fc_1),
          'title_4': 'Frequency response of Low-Pass Filter with fc = {}'.format(fc_2)}

plot(i, normalized_shifted_sinc_1, normalized_shifted_sinc_2, normalized_fft_shifted_sinc_1, 
     normalized_fft_shifted_sinc_2, titles)
```

### Step 2
Spectrally invert the low pass filter with frequency $f_b$.

```python
low_pass = normalized_shifted_sinc_1
high_pass = spectral_inversion(normalized_shifted_sinc_2)

fft_low_pass = normalized_fft_shifted_sinc_1
fft_high_pass = normalize_frequency_function(np.fft.fft(high_pass))

titles = {'title_1':'Sinc - Blackman Window with fc = {}'.format(fc_1),
          'title_2': 'Spectral Inverted Sinc - Blackman Window with fc = {}'.format(fc_2),
          'title_3': 'Frequency response of Low-Pass Filter with fc = {}'.format(fc_1),
          'title_4': 'Frequency response of High-Pass Filter with fc = {}'.format(fc_2)}

plot(i, low_pass, high_pass, fft_low_pass, fft_high_pass, titles)
```

### Step 3
Add both filter kernels.

```python
reject_band = low_pass + high_pass
fft_reject_band = normalize_frequency_function(np.fft.fft(reject_band))
```

### Step 4
Spectral invert the previously generated filter.

```python
band_pass = spectral_inversion(reject_band)
fft_band_pass = normalize_frequency_function(np.fft.fft(band_pass))


titles = {'title_1':'Sinc - Blackman Window with fc1 = {} and fc2 = {}'.format(fc_1, fc_2),
          'title_2': 'Spectral Inverted Sinc - Blackman Window with fc1 = {} and fc2 = {}'.format(fc_1, fc_2),
          'title_3': 'Frequency response of Reject-Band Filter with fc1 = {} and fc2 = {}'.format(fc_1, fc_2),
          'title_4': 'Frequency response of Pass-Band Filter with fc1 = {} and fc2 = {}'.format(fc_1, fc_2)}

plot(i, reject_band, band_pass, fft_reject_band, fft_band_pass, titles)
```
