```python
def shifted_sinc(fc, M):
    """ 
    Function that calculates a sinc shifted time response.
  
    Parameters: 
    fc (float): Cut-off frequency for the low-pass filter. Between 0 and 0.5.
    M (int): Length of the filter kernel. Usually M = 4/BW, where BW is the filter bandwidth of the transition band.
  
    Returns: 
    numpy array: Returns sinc shifted time domain response.
    """
    
    i = (np.arange(0,M,1)).reshape(-1,1)

    t = M//2 + M%2
    limit = np.where(i == t)[0][0]
    h = np.zeros(i.shape)
    h[:limit] = (np.sin(2*np.pi*(i[:limit]-t)*fc))/((i[:limit]-t)*np.pi)
    h[limit+1:] = (np.sin(2*np.pi*(i[limit+1:]-t)*fc))/((i[limit+1:]-t)*np.pi)
    h[limit] = 2*fc
    
    return h
    
    
def hamming_window(M):
    """ 
    Function that calculates a Hamming window of a given M-kernel.
  
    Parameters: 
    M (int): Length of the filter kernel. Usually M = 4/BW, where BW is the filter bandwidth of the transition band.
  
    Returns: 
    numpy array: Returns Hamming window of a given M-kernel.
    """
    i = (np.arange(0,M,1)).reshape(-1,1)
    return 0.54 - 0.46*np.cos(2*np.pi*i/M)

def blackman_window(M):
    """ 
    Function that calculates a Blackman window of a given M-kernel.
  
    Parameters: 
    M (int): Length of the filter kernel. Usually M = 4/BW, where BW is the filter bandwidth of the transition band.
  
    Returns: 
    numpy array: Returns Blackman window of a given M-kernel.
    """
    i = (np.arange(0,M,1)).reshape(-1,1)
    return 0.42 - 0.5*np.cos(2*np.pi*i/M) + 0.08*np.cos(4*np.pi*i/M)


def zero_pad(x, M, L=1024):
    x_padded = np.zeros(L).reshape((-1,1))
    x_padded[0 : M] = x
    return x_padded


def plot_frequency_response(x, f, title="Frequency Response of Filter", label=None):
    # We use this to avoid discontinuities in our plot
    x[0][0] = x[1][0]
    x[-1][0] = x[-2][0]
    
    x_log = 20 * np.log10(x)
    plt.plot(f, x_log, label=label)
    plt.title(title)
    plt.xlabel("Normalized frequency")
    plt.ylabel("Gain [dB]")
    plt.grid("on")
    return
```

```python
def window_filter(fc, M, normalized=True, window_type='hamming'):
    
    sinc = shifted_sinc(fc, M)
    if window_type=='hamming':
        window = hamming_window(M)
    elif window_type=='blackman':
        window = blackman_window(M)
    else:
        raise Exception("unknown window type")
        
    h = sinc*window

    if normalized:
        return h/np.sum(h)
        
    else:
        return h
```

```python
def spectral_reversal(x):
    M = x.shape[0]
    x_reversal = x * ((-1) ** np.arange(M)).reshape(-1,1)

    return x_reversal


def spectral_inversion(x):
    x_inverted = -x
    M = x_inverted.shape[0]
    center = M//2 + M%2
    x_inverted[center] = x_inverted[center] + 1
    return x_inverted
```

```python
fc = 0.20
M = 103
h = window_filter(fc, M, window_type='blackman', normalized=True)
g = window_filter(fc, M, window_type='hamming', normalized=True)
h_inverted = spectral_inversion(h)
g_inverted = spectral_inversion(g)
h_reversal = spectral_reversal(h)
```

```python
def low_pass_filter(fc, M, window_type='blackman', normalized=True):
    return window_filter(fc, M, window_type=window_type, normalized=normalized)

def high_pass_filter(fc, M, method='spectral_inversion',window_type='blackman', normalized=True):
       
    if method=='spectral_inversion':
        h_lp = low_pass_filter(fc, M, window_type=window_type, normalized=normalized)
        h_hp = spectral_inversion(h_lp)
    elif method=='spectral_reversal':
        h_lp = low_pass_filter(0.5-fc, M, window_type=window_type, normalized=normalized)
        h_hp = spectral_reversal(h_lp)
    else:
        raise Exception("method not defined")
        
    return h_hp

def zero_pad_fourier(x, M, L=1024):
    x_padded = zero_pad(x, M, L)
    dft_x = fourier_transform.FourierTransform(x_padded)
    X = dft_x.magx
    f = dft_x.domain
    return X, f
```

```python
fc = 0.2
M = 101
h_lp = low_pass_filter(fc, M)
H_lp, f_lp = zero_pad_fourier(h_lp, M)
```

```python
def band_filter(fc1, fc2, M, band_type='pass', method='spectral_inversion',
                window_type='blackman', normalized=True):
    
    
    if fc1<fc2:
        f1 = fc1
        f2 = fc2
    elif fc1>fc2:
        f1 = fc2
        f2 = fc1
    else:
        raise Exception("fc1 and fc2 cannot be equal") 

        
    h_lp = low_pass_filter(f1, M)
    h_hp = high_pass_filter(f2, M, method=method)
    h = h_lp + h_hp
    
    if band_type=='pass':
        return spectral_inversion(h)
    elif band_type=='reject':
        return h 
    else:
        print("incorrect band_type")
```

```python
fc1 = 0.1
fc2 = 0.2

h = band_filter(fc1, fc2, M, band_type='pass')
H, f = zero_pad_fourier(h, M)

g = band_filter(fc1, fc2, M, band_type='reject')
G, f = zero_pad_fourier(g, M)
```

```python
plot_frequency_response(H, f, label='Band Pass Filter')
plot_frequency_response(G, f, label='Band Reject Filter')
plt.legend()
```
