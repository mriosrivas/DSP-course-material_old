# Convolution

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (5,5)
```

```python
InputSignal_f32_1kHz_15kHz = np.loadtxt(fname = "InputSignal_f32_1kHz_15kHz.dat").flatten()
Impulse_response = np.loadtxt(fname = "Impulse_response.dat").flatten()
```

```python
plt.plot(InputSignal_f32_1kHz_15kHz)
```

```python
plt.plot(Impulse_response)
```

### Single value calculation
$$y[0] = \sum_{j=0}^{M-1}{x[0]h[j]}$$ where $M$ is the lenght  impulse response

```python
output = np.zeros(len(InputSignal_f32_1kHz_15kHz) + len(Impulse_response) - 1)
i = 0 #Between 0 and 28
for j in range(len(InputSignal_f32_1kHz_15kHz)):
    output[i+j] = output[i+j] + Impulse_response[i] * InputSignal_f32_1kHz_15kHz[j]
plt.plot(output)
```

### Convolution
$$\sum_{i=0}^{N-1}{y[i+j]} =  \sum_{j=0}^{M-1}{x[i]h[j]}$$ 

where $M$ is the lenght  impulse response and $N$ the input signal size.

```python
output = np.zeros(len(InputSignal_f32_1kHz_15kHz) + len(Impulse_response) - 1)
for i in range(len(Impulse_response)): 
    for j in range(len(InputSignal_f32_1kHz_15kHz)):
        output[i+j] = output[i+j] + Impulse_response[i] * InputSignal_f32_1kHz_15kHz[j]
plt.plot(output)
```
